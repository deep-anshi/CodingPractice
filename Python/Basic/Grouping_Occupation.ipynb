Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

Step 2. Import the dataset from this address.

```python
data = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep='|')
```

Step 3. Assign it to a variable called users.

```python
users = data
users.head()
```

Step 4. Discover what is the mean age per occupation

```python
#users.groupby(['occupation']).mean().age
#or
users.groupby(['occupation']).age.mean()
```

Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
#users.groupby(['occupation']).count() --> not working

# We have to first change the gender to factor level or numeric
#users["gender_n"] = np.where(users["gender"] == "M", 1, 0)

male_ratio = users.groupby('occupation')['gender_n'].sum() / users.groupby('occupation')['gender_n'].count()
male_ratio.sort_values(ascending=False)  
```

Step 6. For each occupation, calculate the minimum and maximum ages

```python
#users(['min_age']) = users.groupby('occupation').age.min()
#users(['max_age']) = users.groupby('occupation').age.max()
# or  better yet
users.groupby('occupation').age.aggregate(['min', 'max'])

#users.groupby('occupation').age.agg(['min', 'max'])
```

Step 7. For each combination of occupation and gender, calculate the mean age

```python
#users.groupby(['occupation','gender']).age.mean()
users.groupby(['occupation','gender']).mean().age
```

Step 8. For each occupation present the percentage of women and men

```python
#V.imp ques, need to fimd another way/ answer for this (Do it on your own!!)

gender_ocup = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
occup_count = users.groupby(['occupation']).agg('count')
occup_gender = gender_ocup.div(occup_count, level = "occupation") * 100
occup_gender.loc[: , 'gender']

```
